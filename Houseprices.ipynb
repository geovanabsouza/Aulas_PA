```python
# IMPORTAÇÃO DE BIBLIOTECAS
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# CARREGAR OS DADOS
df = pd.read_csv("https://raw.githubusercontent.com/geovanabsouza/Aulas_PA/refs/heads/main/train.csv")
df.head()
```

**1. DEFINIÇÃO DO PROBLEMA**

Qual a questão a ser resolvida? O que esperamos descobrir com a análise?

1.    O tamanho da casa influencia em seu preço?
2.   Como a localização da casa influencia em seu preço?
3.   A quantidade de quartos é um fator influenciável no preço da casa?












```python
# Verificando valores faltantes
df.isnull().sum()
```

```python
# Exibindo informações gerais dos dados
df.info()
```

```python
# Estatísticas descritivas
df.describe()
```

```python
#O Tamanho da casa influencia no preço?

# Gráfico de dispersão
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='LotArea', y='SalePrice', hue='SalePrice', style='SalePrice')
plt.title('Dispersão entre tamanho e preço da casa')
plt.xlabel('Área da casa')
plt.ylabel('Preço')
plt.legend(title='Preço')
plt.show()
```

```python
#Como a localização da casa influecia no seu preço?

# Gráfico de dispersão
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Street', y='SalePrice', hue='SalePrice', style='SalePrice')
plt.title('Dispersão entre localização e preço da casa')
plt.xlabel('Localização da casa')
plt.ylabel('Localização')
plt.legend(title='Preço')
plt.show()
```
